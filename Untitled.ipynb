```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
import random
```

```python
# Base latitude and longitude (example: Mumbai center)
base_lat = 19.0760
base_lon = 72.8777

# Create empty list to hold grid points
grid_data = []

# Loop to create a 10x10 grid
for i in range(10):
    for j in range(10):
        lat = base_lat + i * 0.001   # ~100m increments north
        lon = base_lon + j * 0.001   # ~100m increments east
        noise_db = random.randint(40, 90)  # random noise level in decibels
        grid_data.append([lat, lon, noise_db])

# Convert to a pandas DataFrame
df_grid = pd.DataFrame(grid_data, columns=['latitude', 'longitude', 'noise_db'])

# Show first 5 rows to check
df_grid.head()
```

```python
# Create a map centered near the grid center
m = folium.Map(location=[base_lat + 0.005, base_lon + 0.005], zoom_start=15)

# Prepare heatmap data as list of [lat, lon, weight]
heat_data = [[row['latitude'], row['longitude'], row['noise_db']] for idx, row in df_grid.iterrows()]

# Add heatmap to map
HeatMap(heat_data, radius=15).add_to(m)

# Display map
m
```

```python
import math

# Simulate time of day (hour 0 to 23)
df_grid['hour'] = np.random.randint(0, 24, size=len(df_grid))

# Assume main road is along latitude = base_lat + 0.005 (middle row)
main_road_lat = base_lat + 0.005

# Calculate distance to main road in degrees (approximate)
df_grid['dist_to_main_road'] = abs(df_grid['latitude'] - main_road_lat)

# Now update noise_db to be affected by hour and distance to road
# Noise peaks during rush hours 7-9 AM and 5-7 PM (hours 7-9 and 17-19)

def noise_adjust(row):
    base_noise = row['noise_db']
    hour = row['hour']
    dist = row['dist_to_main_road']
    
    # Increase noise during rush hours by 20 dB, decrease at night by 10 dB
    if 7 <= hour <= 9 or 17 <= hour <= 19:
        base_noise += 20
    elif 22 <= hour or hour <= 5:
        base_noise -= 10
    
    # Noise decreases with distance to main road (scaled by 30)
    base_noise -= dist * 30
    
    # Keep noise between 30 and 100 dB
    return min(max(base_noise, 30), 100)

df_grid['noise_db'] = df_grid.apply(noise_adjust, axis=1)

df_grid.head()
```
